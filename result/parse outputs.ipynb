```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import sys
```

# Plot histgram of read length

```python
d = {}
ds = {}
qscore = []
totlen = []
rep = []
sublen = []
mylen = []
with open("out.fasta") as f:
    for line in f:
        if line[0]=='>':
            li = line.split('_')
            qscore.append(float(li[1]))
            totlen.append(int(li[2]))
            rep.append(int(li[3]))
            sublen.append(int(li[4]))
            mylen.append(int(li[5][:-1]))
            d[li[0][1:]] = li[1:]
        else:
            ds[li[0][1:]] = line
```

```python
tb = pd.DataFrame({'Index':list(range(1,3816))})
tb['Gene'] = d
tb['qscore'] = qscore
tb['totlen'] = totlen
tb['sublen'] = sublen
tb['mylen'] = mylen
tb['rep'] = rep
tb['seq'] = ds.values()
```
